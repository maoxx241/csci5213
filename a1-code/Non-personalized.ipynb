<h2>What are the five movies that most commonly are co-rated with Toy Story</h2>

```python
import numpy as np
def co_rated(df):
    i=1
    count=0
    while True:
        if np.isnan(df[i])==False and np.isnan(table['Toy Story (1995)'][i])==False:
            count=count+1
            
        i = i + 1
        if(i > 610):
            break
    
    return count/table['Toy Story (1995)'].count()
        
```

```python
table.apply(co_rated,axis=0).sort_values(ascending=False)
```

<h2>What are the five movies that have the highest ratings correlations with Toy Story, show movie and correlation?</h2>

```python
def correlation(df):
    return df.corr(table['Toy Story (1995)'],method='pearson')
```

```python
table.apply(correlation,axis=0).sort_values(ascending=False)
```

<h2>the top five movies and this score which shows how much more likely a user is to have rated a movie if they've rated Toy Story. </h2>

```python
def bayes(df):
    i=1
    count=0
    count2=0
    while True:
        if np.isnan(df[i])==False and np.isnan(table['Toy Story (1995)'][i])==False:
            count=count+1
            
        if np.isnan(df[i])==False:
            count2=count2+1
            
        i = i + 1
        if(i > 610):
            break
    
    return (count/610)/((count2/610)*(table['Toy Story (1995)'].count()/610))
```

```python
table.apply(bayes,axis=0).sort_values(ascending=False)
```
